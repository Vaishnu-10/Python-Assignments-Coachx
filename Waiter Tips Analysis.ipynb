```python
import pandas as pd
import altair as alt
```

```python
data=pd.read_csv('Downloads/tips.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
print(data.columns)
```

```python
#1. Histogram of Total Bill
hist_total_bill = alt.Chart(data).mark_bar().encode(
    alt.X('total_bill:Q', bin=True),
    y='count()',
    tooltip=['total_bill', 'count()']
).properties(
    title='Distribution of Total Bill'
)
hist_total_bill
```

```python
#2. Histogram of Tip Amount
hist_tip = alt.Chart(data).mark_bar().encode(
    alt.X('tip:Q', bin=True),
    y='count()',
    tooltip=['tip', 'count()']
).properties(
    title='Distribution of Tip Amount'
)
hist_tip
```

```python
#3. Bar Chart of Tips by Sex
bar_tips_by_sex = alt.Chart(data).mark_bar().encode(
    x='sex:N',
    y='average(tip):Q',
    color='sex:N',
    tooltip=['sex', 'average(tip)']
).properties(
    title='Average Tip Amount by Sex'
)
bar_tips_by_sex
```

```python
#4. Bar Chart of Tips by Smoker
bar_tips_by_smoker = alt.Chart(data).mark_bar().encode(
    x='smoker:N',
    y='average(tip):Q',
    color='smoker:N',
    tooltip=['smoker', 'average(tip)']
).properties(
    title='Average Tip Amount by Smoker'
)
bar_tips_by_smoker
```

```python
#5. Bar Chart of Tips by Day of the Week
bar_tips_by_day = alt.Chart(data).mark_bar().encode(
    x='day:N',
    y='average(tip):Q',
    color='day:N',
    tooltip=['day', 'average(tip)']
).properties(
    title='Average Tip Amount by Day of the Week'
)
bar_tips_by_day
```

```python
#6. Bar Chart of Tips by Time of Day
bar_tips_by_time = alt.Chart(data).mark_bar().encode(
    x='time:N',
    y='average(tip):Q',
    color='time:N',
    tooltip=['time', 'average(tip)']
).properties(
    title='Average Tip Amount by Time of Day'
)
bar_tips_by_time
```

```python
#7. Scatter Plot of Total Bill vs. Tip
scatter_total_bill_tip = alt.Chart(data).mark_circle().encode(
    x='total_bill:Q',
    y='tip:Q',
    color='sex:N',
    tooltip=['total_bill', 'tip', 'sex']
).properties(
    title='Total Bill vs. Tip Amount'
)
scatter_total_bill_tip
```

```python
#8. Scatter Plot of Total Bill vs. Tip, Colored by Smoker
scatter_total_bill_tip_smoker = alt.Chart(data).mark_circle().encode(
    x='total_bill:Q',
    y='tip:Q',
    color='smoker:N',
    tooltip=['total_bill', 'tip', 'smoker']
).properties(
    title='Total Bill vs. Tip Amount (Colored by Smoker)'
)
scatter_total_bill_tip_smoker
```

```python
#9. Box Plot of Total Bill by Day
box_total_bill_day = alt.Chart(data).mark_boxplot().encode(
    x='day:N',
    y='total_bill:Q',
    color='day:N',
    tooltip=['day', 'total_bill']
).properties(
    title='Total Bill by Day of the Week'
)
box_total_bill_day
```

```python
#10. Heatmap of Total Bill by Day and Time
heatmap_total_bill_day_time = alt.Chart(data).mark_rect().encode(
    x='day:N',
    y='time:N',
    color='average(total_bill):Q',
    tooltip=['day', 'time', 'average(total_bill)']
).properties(
    title='Average Total Bill by Day and Time'
)
heatmap_total_bill_day_time
```
