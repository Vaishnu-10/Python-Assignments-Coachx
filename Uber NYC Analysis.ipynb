```python
import pandas as pd
import altair as alt
```

```python
data=pd.read_csv('Downloads/Uber.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
print(data.columns)
```

```python
#1. Histogram of Active Vehicles
hist_active_vehicles = alt.Chart(data).mark_bar().encode(
    alt.X('active_vehicles:Q', bin=True),
    y='count()',
    tooltip=['active_vehicles', 'count()']
).properties(
    title='Distribution of Active Vehicles'
)
hist_active_vehicles
```

```python
#2. Histogram of Trips
hist_trips = alt.Chart(data).mark_bar().encode(
    alt.X('trips:Q', bin=True),
    y='count()',
    tooltip=['trips', 'count()']
).properties(
    title='Distribution of Trips'
)
hist_trips
```

```python
#3. Bar Chart of Average Trips by Dispatching Base Number
bar_trips_by_base = alt.Chart(data).mark_bar().encode(
    x='dispatching_base_number:N',
    y='average(trips):Q',
    color='dispatching_base_number:N',
    tooltip=['dispatching_base_number', 'average(trips)']
).properties(
    title='Average Trips by Dispatching Base Number'
)
bar_trips_by_base
```

```python
#4. Line Chart of Trips Over Time
line_vehicles_over_time = alt.Chart(data).mark_line().encode(
    x='date:T',
    y='active_vehicles:Q',
    tooltip=['date', 'active_vehicles']
).properties(
    title='Active Vehicles Over Time'
)
line_vehicles_over_time

```

```python
#6. Bar Chart of Average Active Vehicles by Dispatching Base Number
bar_vehicles_by_base = alt.Chart(data).mark_bar().encode(
    x='dispatching_base_number:N',
    y='average(active_vehicles):Q',
    color='dispatching_base_number:N',
    tooltip=['dispatching_base_number', 'average(active_vehicles)']
).properties(
    title='Average Active Vehicles by Dispatching Base Number'
)
bar_vehicles_by_base
```

```python
#7. Scatter Plot of Active Vehicles vs. Trips
scatter_vehicles_trips = alt.Chart(data).mark_circle().encode(
    x='active_vehicles:Q',
    y='trips:Q',
    color='dispatching_base_number:N',
    tooltip=['active_vehicles', 'trips', 'dispatching_base_number']
).properties(
    title='Active Vehicles vs. Trips'
)
scatter_vehicles_trips
```

```python
#8. Box Plot of Trips by Dispatching Base Number
box_trips_by_base = alt.Chart(data).mark_boxplot().encode(
    x='dispatching_base_number:N',
    y='trips:Q',
    color='dispatching_base_number:N',
    tooltip=['dispatching_base_number', 'trips']
).properties(
    title='Trips by Dispatching Base Number'
)
box_trips_by_base
```
