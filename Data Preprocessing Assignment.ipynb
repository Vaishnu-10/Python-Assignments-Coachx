```python
'''3)Write a Python program for Handling Missing Value. Replace missing value of salary, age column with mean of that column.(Use Data.csv file).'''
import pandas as pd
df=pd.read_csv('Downloads/Salary_Data.csv')
missing_values=df.isnull().sum()
print(missing_values)
mean_salary=df['Salary'].mean()
mean_age=df['Age'].mean()
df['Salary'].fillna(mean_salary,inplace=True)
df['Age'].fillna(mean_age,inplace=True)
missing_values=df.isnull().sum()
print(missing_values)
df.to_csv('Downloads/Salary_Data_modified.csv', index=False)
print(df)
```

```python
'''3)Generate a random array of 50 integers and display them using a line chart, scatter plot, histogram and box plot. Apply appropriate color, labels and styling options.'''
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randint(0, 100, size=50)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
# Line Chart
axs[0, 0].plot(data, color='blue', marker='o', linestyle='-', markersize=5)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Value')
axs[0, 0].grid(True)
# Scatter Plot
axs[0, 1].scatter(range(len(data)), data, color='red', marker='x')
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('Index')
axs[0, 1].set_ylabel('Value')
axs[0, 1].grid(True)
# Histogram
axs[1, 0].hist(data, bins=10, color='green', edgecolor='black')
axs[1, 0].set_title('Histogram')
axs[1, 0].set_xlabel('Value')
axs[1, 0].set_ylabel('Frequency')
axs[1, 0].grid(True)
# Box Plot
axs[1, 1].boxplot(data, patch_artist=True, boxprops=dict(facecolor='cyan', color='blue'))
axs[1, 1].set_title('Box Plot')
axs[1, 1].set_ylabel('Value')
axs[1, 1].grid(True)
plt.tight_layout()
plt.show()
```

```python
'''Assignment  -4 :-

1)Write a Python program to perform the following tasks : 
a.Apply OneHot coding on Country column. 
b. Apply Label encoding on purchased column (Data.csv have two categorical column the country column, and the purchased column)
'''
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
df = pd.read_csv('Downloads/Data.csv')
onehotencoder = OneHotEncoder()
country_encoded = onehotencoder.fit_transform(df[['Country']]).toarray()
country_encoded_df = pd.DataFrame(country_encoded, columns=onehotencoder.get_feature_names_out(['Country']))
df = pd.concat([df, country_encoded_df], axis=1).drop('Country', axis=1)
labelencoder = LabelEncoder()
df['Purchased'] = labelencoder.fit_transform(df['Purchased'])
print(df.head())
```

```python
''' 2)Write a python program to compute sum of Manhattan distance between all pairs of points.'''
def manhattan_distance(point1, point2):
    return abs(point1[0] - point2[0]) + abs(point1[1] - point2[1])
def total_manhattan_distance(points):
    total_distance = 0
    num_points = len(points)
    
    for i in range(num_points):
        for j in range(i + 1, num_points):
            total_distance += manhattan_distance(points[i], points[j])
    
    return total_distance
points = [(1, 2), (3, 4), (6, 1), (5, 3)]
result = total_manhattan_distance(points)
print("The total Manhattan distance between all pairs of points is:", result)
```

```python
'''4)Write a Python NumPy program to compute the weighted average along the specified axis of a given flattened array.'''
import numpy as np

def compute_weighted_average(arr, weights, axis=None):
    if axis is not None:
         weights = np.broadcast_to(weights, arr.shape)
    weighted_sum = np.sum(arr * weights, axis=axis)
    sum_of_weights = np.sum(weights, axis=axis)
    return weighted_sum / sum_of_weights
arr = np.array([[1, 2, 3], [4, 5, 6]])
weights = np.array([0.1, 0.2, 0.7])
axis = 1  
result = compute_weighted_average(arr, weights, axis=axis)
print("The weighted average along the specified axis is:", result)
```

```python
'''5)Write a python program to create two lists, one representing subject names and the other representing marks obtained in those subjects. Display the data in a pie chart and bar chart'''
import matplotlib.pyplot as plt

subjects = ['Math', 'Science', 'English', 'History', 'Art']
marks = [85, 90, 78, 88, 92]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.pie(marks, labels=subjects, autopct='%1.1f%%', startangle=140)
plt.title('Marks Distribution in Subjects')

plt.subplot(1, 2, 2)
plt.bar(subjects, marks, color='skyblue')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Obtained in Subjects')

plt.tight_layout()
plt.show()
```
