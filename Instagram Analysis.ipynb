```python
import pandas as pd
import altair as alt
```

```python
file_path = 'Downloads/Instagram_data.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display the first few rows of the data
data.head()
```

```python
data.describe()
```

```python
#1. Histogram of Impressions
hist_impressions = alt.Chart(data).mark_bar().encode(
    alt.X('Impressions', bin=True),
    y='count()',
).properties(
    title='Histogram of Impressions'
)

hist_impressions
```

```python
#2. Bar Chart for Impressions from Different Sources
bar_sources = alt.Chart(data).transform_fold(
    ['From Home', 'From Hashtags', 'From Explore', 'From Other'],
    as_=['Source', 'Impressions']
).mark_bar().encode(
    x='Source:N',
    y='sum(Impressions):Q',
    color='Source:N'
).properties(
    title='Impressions from Different Sources'
)

bar_sources
```

```python
#3. Grouped Bar Chart for Engagement Metrics
bar_engagement = alt.Chart(data).transform_fold(
    ['Saves', 'Comments', 'Shares', 'Likes', 'Profile Visits', 'Follows'],
    as_=['Metric', 'Value']
).mark_bar().encode(
    x='Metric:N',
    y='sum(Value):Q',
    color='Metric:N',
    column='Metric:N'
).properties(
    title='Engagement Metrics'
)

bar_engagement
```

```python
#4. Scatter Plot for Likes vs. Profile Visits
scatter_likes_profile = alt.Chart(data).mark_point().encode(
    x='Likes:Q',
    y='Profile Visits:Q',
    color='Profile Visits:Q'
).properties(
    title='Scatter Plot of Likes vs. Profile Visits'
)

scatter_likes_profile
```

```python
#5. Line Chart for Impressions Over Index
line_impressions = alt.Chart(data.reset_index()).mark_line().encode(
    x='index:Q',
    y='Impressions:Q'
).properties(
    title='Line Chart of Impressions Over Time'
)

line_impressions
```

```python
#6. Heat Map for Saves vs. Likes with Scatter Plot Overlay
heatmap_saves_likes = alt.Chart(data).mark_rect().encode(
    x='Saves:Q',
    y='Likes:Q',
    color='count()'
).properties(
    title='Heat Map of Saves vs. Likes'
)

scatter_saves_likes = alt.Chart(data).mark_point().encode(
    x='Saves:Q',
    y='Likes:Q',
    size='Likes:Q',
    color='Likes:Q'
)

heatmap_saves_likes + scatter_saves_likes
```
